```python
from zoneinfo import ZoneInfo
from datetime import datetime

def convert_barset_to_LA(barset):
    la_tz = ZoneInfo("America/Los_Angeles")
    for bar in barset:
        bar.timestamp = bar.timestamp.astimezone(la_tz)
    return barset
```
